## 공영주차장 설치 제안 장소 선별
2022년, 2023년의 주정차위반 예측 결과 중 증가폭이 큰 75개 지역을 각 각 선별한 후, <br>
이를 하나로 합치고 주차장 설치 여부를 종합적으로 판단하여 <br>
30개의 구역을 최종적으로 선별

```python
df_suggest = pd.read_csv('./df_candidate.csv')

#좌표를 기준으로 구역을 다시 설정
for i in range(len(df_suggest)):
    for j in range(len(df15['features'])):
        if Point(df_suggest[['long', 'lat']].iloc[i]).within(Polygon(df15['features'][j]['geometry']['coordinates'][0])) == True:
            df_suggest.loc[i, 'id'] = df15['features'][j]['properties']['id'] 

df_suggest = df_suggest.drop_duplicates(['id'])
df_suggest
```

```python
m = folium.Map(location=[35.126703, 129.100194], zoom_start=13, prefer_canvas=True)

folium.GeoJson(df15).add_to(m)
for i in range(len(df_suggest)):
    folium.Circle(location = [df_suggest['lat'].iloc[i], df_suggest['long'].iloc[i]], radius=100, 
                  color='green', fill_color='green', fill_opacity=0.3, line_opcity=0.1, tooltip = df_suggest['id'].iloc[i]).add_to(m)
m
```

## 기존 주차장 활용이 가능한 구역 선별
기존에 존재하는 공영주차장의 병합 또는 증설 가능한 구역을 선별 <br>
빨간점은 설치제안장소, 초록색은 기존에 존재하는 공영주차장의 위치

```python
m = folium.Map(location=[35.126703, 129.100194], zoom_start=13, prefer_canvas=True)

keys = df_suggest['id']
values = df_suggest['how_many']
temp = pd.DataFrame({'id' : keys, 'count' : values})

folium.Choropleth(geo_data=df15, data=temp, columns=('id', 'count'), key_on='feature.properties.id', 
                  fill_color='YlGn',legend_name='주정차 위반',).add_to(m)
for i in range(len(df3)):
    folium.Circle(location = [df3['lat'].iloc[i], df3['lon'].iloc[i]], radius=float(df3['park_slot'].iloc[i]*2), 
                  color='green', fill_color='green', fill_opacity=0.3, line_opcity=0.1, tooltip = df3['location'].iloc[i]).add_to(m)
    
for i in range(len(df_suggest)):
    folium.Circle(location = [df_suggest['lat'].iloc[i], df_suggest['long'].iloc[i]], radius=float(df_suggest['increasing'].iloc[i])*5, 
                  color='red', fill_color='red', fill_opacity=0.3, line_opcity=0.1, tooltip = df_suggest['id'].iloc[i]).add_to(m)   
m
```

```python
temp = [223,400,541,654,506,545,402,576,406,434,650,392,683,294,728,651,356,291,427,622,329,579,399,390,228,421,304,469,581,149]
df_suggest = df_suggest[[i for i in df_suggest['id'].isin(temp)]].reset_index(drop=True)

temp = ['인근주택매입','인근주택매입','인근주택매입','인근주택매입','인근부지매입','기존주차장병합','인근부지매입','인근주택매입',
        '기존주차장증설','인근주택매입','기존주차장증설','인근부지매입','인근주택매입','인근주택매입','인근부지매입','인근주택매입',
        '인근부지매입','기존주차장병합','인근주택매입','기존주차장증설','인근주택매입','인근주택매입','인근주택매입','인근부지매입',
        '인근부지매입','인근부지매입','인근부지매입','기존주차장증설','기존주차장병합','인근주택매입']
df_suggest['how'] = temp

temp = ['-','-','-','-','-','용호1동 72번지 용호1동 420-12 용호1동 423','-','-','용당동 136-3번지','-','용호3동 565번지',
        '-','-','-','-','-','-','대연6동 1801 대연6동 1541-5번지 대연5동 1535-2','-','용호동 895-5번지','-','-','-','-','-',
        '-','-','대연3동 1022','용호1동 72번지 용호1동 420-12 용호1동 423','-']
df_suggest['where'] = temp
  
temp = []
for i in range(len(df_suggest)):
    temp.append(df_1[df_1['id'] == df_suggest['id'].iloc[i]]['cd_loc'].iloc[0])
df_suggest['cd_loc'] = temp
df_suggest = df_suggest.drop(['age', 'price'], axis = 1)
df_id = df_suggest

df_suggest['설치순위'] = range(1,31)
#구역을 기준으로 좌표를 해당 구역의 중심점으로 다시 설정
for i in range(len(df_suggest)):
    for j in range(len(df15['features'])):
        if df_suggest['id'].iloc[i] == df15['features'][j]['properties']['id']:
            df_suggest.loc[i, 'X좌표'] = Polygon(df15['features'][j]['geometry']['coordinates'][0]).centroid.x
            df_suggest.loc[i, 'Y좌표'] = Polygon(df15['features'][j]['geometry']['coordinates'][0]).centroid.y
            
df_suggest.rename(columns = {'cd_loc':'주차장 위치', 'where':'기존 주차장 위치', 'how_many':'주차용량', 'how':'비고'}, inplace=True)

# 주차장 면적 계산(차량 한대의 주차 면적은 가로 2.5 세로 5.0)
temp = []
for i in range(len(df_suggest)):
    temp.append(int((df_suggest['주차용량'].iloc[i])*2.5*5.0))
df_suggest['주차장 면적'] = temp
df_suggest = df_suggest[['설치순위', '주차장 위치', 'X좌표', 'Y좌표', '주차장 면적', '주차용량', '비고']]
df_suggest.to_csv('df_suggest.csv', index=False)
df_suggest
```
